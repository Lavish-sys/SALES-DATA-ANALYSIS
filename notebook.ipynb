# SALES PERFORMANCE & PRODUCT INSIGHT ANALYSIS

This project analyzes sales data to understand product performance, category trends, and regional distribution. The goal is to identify high performing and low performing areas to support better business decisions.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")

df.head()
```

```python
df.columns
```

```python
df.info()
df.describe()
```

```python
df.groupby("Category")["Sales"].sum()
```

```python
df.groupby("Sub-Category")["Sales"].sum().sort_values()
```

```python
df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby("Product Name")["Sales"].sum().sort_values().head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Category", y="Sales", data=df)
plt.show()
```

This graph shows how sales are distributed across categories. It helps identify which category contributes the most to overall revenue.

```python
sns.barplot(x="Region", y="Sales", data=df)
plt.show()
```

This graph highlights sales performance across different regions. Some regions clearly perform better than others.

```python
df.groupby("Sub-Category")["Sales"].sum().plot(kind="bar")
plt.show()
```

This graph shows variation in sales across subcategories. Some products perform significantly better than others.

# KEY INSIGHTS
  --> Some categories contribute significantly more to total sales than others.  
  --> Certain subcategories have very low sales, indicating weak performance.  
  --> Sales vary across regions, with some regions performing better.  
  --> A small number of products generate most of the revenue.


## CONCLUSION
The analysis shows that businesses should focus on high performing products while identifying and improving low performing ones. Understanding regional trends can also help improve overall sales strategy.
